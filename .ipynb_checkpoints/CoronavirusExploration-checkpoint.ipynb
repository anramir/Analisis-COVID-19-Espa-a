```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import urllib.request as request
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
FILES_TO_DOWNLOAD = [
    'https://covid.ourworldindata.org/data/total_cases.csv',
    'https://covid.ourworldindata.org/data/total_deaths.csv',
    'https://covid.ourworldindata.org/data/new_cases.csv',
    'https://covid.ourworldindata.org/data/new_deaths.csv'
]
```

```python
for file in FILES_TO_DOWNLOAD:
    file_name = file.split("/")[-1]
    request.urlretrieve(file, f'data/{file_name}')
```

```python
df_total_cases = pd.read_csv('data/total_cases.csv').set_index('date').transpose().fillna(0)
df_total_cases.loc['Spain']
```

```python
fig, ax = plt.subplots(1, figsize = (10, 5))
sns.heatmap(df_total_cases)
```

```python
def plot_regression(df, country, prediction_days = 7, title = "Values"):
    fig, ax = plt.subplots(1, figsize = (15, 5))
    ax.set_yscale('log')
    filtered = df.loc[country].where(lambda x: x > 10).dropna()
    log = filtered.map(lambda x: np.log10(x))
    reg = LinearRegression().fit(X = np.arange(len(log)).reshape(-1, 1), y = log)
    X = np.arange(len(log)).reshape(-1, 1)
    R_2 = reg.score(X = X, y = log)
    print(f"R_2 : {R_2}")
    print(f"Params : {reg.coef_}")
    print(f"Every {1 / reg.coef_[0] } days, quantity is multiplied by 10")
    time_range = pd.date_range(start = filtered.index.tolist()[0], end = pd.Timestamp(filtered.index.tolist()[-1]) + pd.DateOffset(days = prediction_days))
    X_with_days_prediction = np.arange(len(time_range)).reshape(-1, 1)
    prediction = np.array(reg.predict(X_with_days_prediction))
    df_prediction = pd.DataFrame(data = prediction, columns = ['prediction'])
    df_prediction.index = time_range
    df_prediction.prediction = df_prediction.prediction.map(lambda x: np.power(10, x))
    df_prediction.plot(ax = ax)
    filtered = pd.DataFrame(filtered)
    filtered.index = filtered.index.map(lambda x: pd.to_datetime(x))
    filtered.plot(ax = ax)
    ax.grid()
    ax.set_title(f'{title} in {country}')
    ax.legend()
    ax.grid(which='major', color='#CCCCCC', linestyle='--')
    ax.grid(which='minor', color='#CCCCCC', linestyle=':')
    print(df_prediction.tail(prediction_days))
```

```python
plot_regression(df_total_cases, 'Spain', 11, title = "Total COVID-19 infected")
```

```python
df_new_cases = pd.read_csv('data/new_cases.csv').set_index('date').transpose().fillna(0)
df_total_cases
```

```python
plot_regression(df_new_cases, 'Spain', 11, title = "New cases")
```

```python
df_new_deaths = pd.read_csv('data/new_deaths.csv').set_index('date').transpose().fillna(0)
df_new_deaths
```

```python
plot_regression(df_new_deaths, 'Spain')
```

```python
df_total_deaths = pd.read_csv('data/total_deaths.csv').set_index('date').transpose().fillna(0)
df_total_deaths
```

```python
plot_regression(df_total_deaths, 'Spain', 11, title = "Deaths")
```
